```python
import pandas as pd
import networkx as nx

categories_df = pd.read_csv("categories.csv")
category_map = dict(zip(categories_df["CategoryId"], categories_df["names"]))

coms = nx.community.louvain_communities(G, weight='weight', resolution=1, threshold=1e-07, max_level=None, seed=None)

# Replace node IDs in each community with their category names
coms_with_category_names = []
for community in coms:
    community_names = set()
    for node in community:
        # Split categories and map each to its name, ignoring empty values
        category_numbers = str(G.nodes[node]["categories"]).split(',')
        category_names = {
            category_map[int(cat.strip())]
            for cat in category_numbers
            if cat.strip() and int(cat.strip()) in category_map  # Ensure cat is non-empty and in the map
        }
        community_names.update(category_names)
    coms_with_category_names.append(community_names)

coms_with_category_names
```
