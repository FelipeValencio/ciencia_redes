### 1- Visualize a rede “GraphMissingEdges.gml” (referente ao desafio no Kaggle) com o layout ForceAtlas2.

 Arquivo ex1_view_atlas2.png

```python
import networkx as nx
import matplotlib.pyplot as plt

N = 3000  # Número de nós
p = 10**-3  # Probabilidade de conexão entre dois nós

G = nx.erdos_renyi_graph(N, p)

# Informações da rede
num_nos = G.number_of_nodes()
num_links = G.number_of_edges()

links_esperados = (N * (N - 1)) / 2 * p


print(f"Número de nós: {num_nos}")
print(f"Número de links: {num_links}")
print(f"Links esperados pela fórmula: {links_esperados}")

### a) Qual o número de links esperado? Isso vai de acordo com o que foi observado na rede criada?
# Aplicando a fórmula de número médio de links em um grafo aleatório temos como resultado 4498.5
# Podemos ver que os valores reais são próximos do esperado, variando aproxidamente 2.5% da expectativa para mais e menos
```

### b) Em qual regime essa rede se encontra, subcritical, critical, supercritical ou conected? Justifique.

Precisamos fazer algumas análises com as fórmulas 1/N e ln(N)/N. Com o código abaixo vemos que a rede é SuperCritical pois o valor de p (10^-3) é superior a 1/N (1/3000)

```python
import math

print(f"Subcritical? {p < 1/N}")
print(f"Critical? {p == 1/N}")
print(f"SuperCritical? {p > 1/N}")
print(f"Conectada? {p >= math.log(N)/N}")
```

### 3- Com base na rede fornecida para o desafio, arquivo “GraphMissingEdges.gml”, faça:
a) Compute a distribuição de graus.

b) Compute a distribuição de graus para duas redes aleatórias com o mesmo número de nós e não direcionadas, mas uma com p=0,001 e outra com p =0,01.

c) A rede do desafio possui características de redes aleatórias? Justifique a sua resposta.

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.read_gml("GraphMissingEdges.gml")

# Calcula a distribuição de graus
degrees = [deg for node, deg in G.degree()]
plt.hist(degrees, bins=30, edgecolor="black", alpha=0.7)
plt.xlabel("Grau")
plt.ylabel("Frequência")
plt.title("Distribuição de Graus - Rede do Desafio")
plt.show()
```

```python
# Número de nós na rede do desafio
N = G.number_of_nodes()

# Rede com p =0.001
p1 = 0.001
G_random_1 = nx.erdos_renyi_graph(N, p1)
degrees_random_1 = [deg for node, deg in G_random_1.degree()]
plt.hist(degrees_random_1, bins=30, edgecolor="black", alpha=0.7)
plt.xlabel("Grau")
plt.ylabel("Frequência")
plt.title("Distribuição de Graus - Rede Aleatória (p = 0.001)")
plt.show()

# Rede com p =0.01
p2 = 0.01
G_random_2 = nx.erdos_renyi_graph(N, p2)
degrees_random_2 = [deg for node, deg in G_random_2.degree()]
plt.hist(degrees_random_2, bins=30, edgecolor="black", alpha=0.7)
plt.xlabel("Grau")
plt.ylabel("Frequência")
plt.title("Distribuição de Graus - Rede Aleatória (p = 0.01)")
plt.show()
```

### c) A rede do desafio possui características de redes aleatórias? Justifique a sua resposta.

Não, podemos perceber pela distribuição de graus das redes aleatórias que a distribuição é muito mais balanceada do que a da rede do desafio, onde existe uma gigantesca concentração de nós com grau baixo e que diminui conforme a quantidade de graus aumenta. 
Também podemos comparar a distribuição dos graus do grafos com a distribuição de Poisson. É claramente perceptível que a distribuição dos grafos aleatórios é muito próximo a distribuição de Poisson, enquanto a da rede do desafio se assimila com de outras redes reais como da internet por exemplo.
