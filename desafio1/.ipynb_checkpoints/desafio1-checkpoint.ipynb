```python
import networkx as nx
import matplotlib.pyplot as plt
import scipy.sparse
G = nx.Graph()
```

```python
G = nx.read_gml("GraphMissingEdges.gml")

print(list(G.nodes))
print(list(G.edges))
print(G.number_of_nodes())
```

### Grau dos nós

```python
G.degree()
```

### Visualizando a distribuição de graus

```python
degree_sequence = [d for n, d in G.degree()]
plt.hist(degree_sequence, bins=50)
plt.title("Distribuição de grau")
plt.xlabel("Grau")
plt.ylabel("Conta")
plt.show()

# Essa visualização é interessante pois nos mostra que existem milhares de locais sem nenhuma conexão (grau 0) e poucos com alto nível de grau.
# Podemos já concluir que dentro desse dataset locais que obtiveram avaliações do mesmo usuário são relativamente raros
```

```python
# Create a new graph without nodes that have degree 0
nodes_with_degree_0 = [node for node, degree in G.degree() if degree == 0]
G_filtered = G.copy()  # Make a copy of the original graph
G_filtered.remove_nodes_from(nodes_with_degree_0)  # Remove degree 0 nodes

# Verify the filtering
print(f"Original graph nodes: {G.number_of_nodes()}")
print(f"Filtered graph nodes: {G_filtered.number_of_nodes()}")


# Get degree sequence from the filtered graph
degree_sequence_filtered = [degree for node, degree in G_filtered.degree()]

print(degree_sequence_filtered)

# Plot the degree distribution for the filtered graph
plt.hist(degree_sequence_filtered, bins=50)
plt.title("Degree Distribution (excluding nodes with degree 0)")
plt.xlabel("Degree")
plt.ylabel("Count")
plt.show()
```

### Coeficiente de agrupamento / clustering coeficient

```python
clustering_dict = nx.clustering(G)

sorted_clustering = sorted(clustering_dict.items(), key=lambda x: x[1], reverse=True)

print(sorted_clustering)

# Este número pode ser interessante para esse problema, por exemplo ao saber que um node tem um alto número de clusterização
# podemos supor que seus vizinhos também estão conectadas entre si. Esse indice pode ser uma das principais ferramentas para fazer ou checar
# a previsão de conexões entre locais
```

```python
print(nx.average_clustering(G))

# E com o valor médio de clusterização em 0.0448... podemos ver que existe um baixo nível pelo grafo 
```
